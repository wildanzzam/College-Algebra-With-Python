### Demand and Revenue

#### Selling Tee Shirts Example

```python
price = 10
demand = 50 - 2*price 
revenue = price*demand 
total_cost = 4*demand 
profit = revenue - total_cost 

print("price: ", price)
print("demand: ", demand)
print("revenue: ", revenue)
print("total cost: ", total_cost)
print("profit ", profit)
```

#### Supply and Demand Graph

```python
import matplotlib.pyplot as plt

old_demand_x = 1
new_demand_x = 12
old_price_y = 14
new_price_y = 3
old_supply_x = 11
new_suppy_x = 1

# For the graph
xmin = 0 
xmax = 15
ymin = 0
ymax = 15

# Basic setup for the graph
fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # Window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax],'b') # blue y axis
plt.xlabel("Quantity")
plt.ylabel("Price")
plt.title("Green=supply, Red=Demand")

# Plot the demand function as a red line
plt.plot([old_demand_x, new_demand_x], [old_price_y,new_price_y], 'r')

# Plot the supply funtion as a green line
plt.plot([old_supply_x,new_suppy_x],[old_price_y,new_price_y],'g')

plt.show()
```

`Demand`

- If the price is high (14), then the quantity that is demanded by market is low (1)
- When the price is low (3), the quantity that is demanded by market is high (12)

`Supply`
- When the price is low (3), then the owner of business won't jump into this business as the demand is low (1)
- When the price is high (14), then the owner of business will jump into industry as the demand is high (11)

`èquilibrium`
- Where these lines intersect
- Quantity supply and demand are same

Basically, the customers want the cheapest price and more commodity. While the business want to generate more profit by selling more items.

#### Calculate the Price Elasticity of Demand

$ \frac{percent \; change \; in \; demand}{percent \; change \; in \; price} $

elasticity is how quickly does demand change when the price changes

```python
old_demand = 5
new_demand = 6
old_price = 11
new_price = 10

# percent change in price
price_change = (new_price - old_price) / old_price

# percent change in demand
demand_change = (new_demand - old_demand) / old_demand

e_number = demand_change/price_change

# elasticity
print(e_number)

if e_number>1:
    print("Demand is elastic")
elif e_number==1:
    print("Demand is unitary or proportional")
elif e_number<1:
    print("Demand is inelastic")
else:
    print("?")
```

#### Calculate Price Elasticity of Supply

$ \frac{percent \; change \; in \; supply}{percent \; change \; in \; price} $

how responsive the supply with the changes of prices

```python
old_price = 4
new_price = 7
old_supply = 10
new_supply = 11

# percent change in price
price_change = (new_price - old_price) / old_price

# percent change in supply
supply_change = (new_supply - old_supply) / old_supply

e_number = supply_change/price_change

# elasticity
print(e_number)

if e_number>1:
    print("Supply is elastic")
elif e_number==1:
    print("Supply is unitary or proportional")
elif e_number<1:
    print("Supply is inelastic")
else:
    print("?")
```
