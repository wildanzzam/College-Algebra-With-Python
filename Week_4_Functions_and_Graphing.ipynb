### Functions and Graphing

#### Function and a table of (x, y) values

```python
x = 5
y = 4*x + 3 # The function 

print(x,",",y)

print("x \t y") # defining column

for x in range(11): # Loop
    y = 4*x + 3  # The function
    print(x, "\t", y) # put the tab between
```

#### Defining a Python function

```python
# Finding the y
def f(x): # Define the function
   y = 4*x + 3 # indented function
   return y 

print(5, ",", f(5)) 

# Next, a loop
for x in range(11):
    # y = 4*x + 3  # The function already defined
    print(x, "\t", f(x)) 
```

#### Display Axis Lines

```python
import matplotlib.pyplot as plt

xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
                                # It will draw the line from point 0 and 0
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

plt.show()
```

#### Plot Several Points as a Function

```python
import matplotlib.pyplot as plt

xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

# Loop 
for x in range(10):
    y = 0.5*x + 1
    plt.plot([x],[y], 'ro')

plt.show()
```

#### Graph and Table of (x, y) values

```python
import matplotlib.pyplot as plt

xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

print("x \t y")
for x in range(xmin, xmax+1): #xmax + 1 because last point does not include the last number
                              # so we + 1 in order for the number 10 to be included.
    y = 0.5*x + 1
    plt.plot([x],[y], 'ro')
    print(x,"\t",y)

plt.show()
```

### Extra Problems

#### Loop Through Function Inputs

```python
import matplotlib.pyplot as plt

xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

for x in range(10):
    
    y = 0.5*x + 1
    plt.plot([x],[y], 'ro')

plt.show()
```

#### Use an Array as Inputs

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2*(xmax-xmin) 
x = np.linspace(xmin, xmax, points) #linspace function to set up our array

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

y = 2*x +1
plt.plot(x,y, 'green')

plt.show()
```

#### More Ways to Customize a Graph

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2*(xmax-xmin)
x = np.linspace(xmin, xmax, points)

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

ax.set_xlabel("x values")
ax.set_ylabel("y values")
ax.set_title("Some Graph")
#ax.grid(True)

ax.set_xticks(np.arange(xmin, xmax, 1))
ax.set_yticks(np.arange(ymin, ymax, 1))

y = 2*x +1
plt.plot(x,y, label='y=2x+1')
plt.plot([4],[6], 'ro', label='point')
plt.plot(x,3*x, label='steeper line')
# plt.legend()
plt.show()
```
