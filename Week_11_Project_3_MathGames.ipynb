Scatter plot game

Randomly generate points on a graph and the player has to input the (x,y) coordinates
For added difficulty, make the graph larger

```python
import matplotlib.pyplot as plt
import random

# Step 1: Generate Random Points
num_points = 5
points = [(random.randint(-100, 100), random.randint(-100, 100)) for _ in range(num_points)]
print("Generated Points (for reference):", points) # For reference to check correctness

# Step 2: Plot the Points
x_values = [point[0] for point in points]
y_values = [point[1] for point in points]

# Plot the graph
# Set the dimensions
def f(zoom):
    xmin = -zoom
    xmax = zoom
    ymin = -zoom
    ymax = zoom
    # Make the figure
    fig, ax = plt.subplots()
    # make scatter coordinate
    ax.scatter(x_values, y_values)
    
    # Draw the x-axis and y-axis lines
    # Make the horizontal line
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)

    # Draw lines connecting each point to the origin
    for x, y in points:
        ax.plot([0, x], [0, y], color='green', linestyle='--', linewidth=0.5)

    # Set the limit
    ax.set_xlim([xmin, xmax])
    ax.set_ylim([ymin, ymax]) 
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Scatter Plot Game')
    plt.grid(True)
    plt.show()

# Call the function to display the plot with desired Zoom level
f(100) # You can adjust the zoom level here

# Step 3: Take User Input and Check
score = 0

# enumerate to loop through each point in the points while also keeping track of index 'i'
for i, point in enumerate(points):
    user_input = input(f"Enter the coordinates for point {i+1} (format: x,y):")
    # Apply a given function to all items in an iterable and return an iterator
    user_x, user_y = map(int, user_input.split(','))

    if (user_x, user_y) == point:
        print("Correct!")
        score += 1
    else:
        print(f"Wrong! The correct coordinates were {point}")

print(f"Your final score is {score}/{num_points}")

```
