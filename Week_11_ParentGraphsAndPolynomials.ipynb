#### Sliders

```python
# sliders and zoom
# y = a

# It's a wedget here
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

# Define the graphing function
def f(a, zoom): # a for the slider
    xmin = -zoom
    xmax = zoom
    ymin = -zoom
    ymax = zoom
    # Number of point it takes between xmax and xmin
    points = 10*xmax-xmin
    # Generating array and points between xmin and xmaxc
    x = np.linspace(xmin, xmax, points)

    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0],'black') # black x axis
    plt.plot([0,0],[ymin,ymax], 'black') # black y axis
    
    # Line 1
    y1 = a
    y2 = a
    plt.plot([xmin,xmax], [y1,y2]) # Plug horizontal line
    plt.show()

# Set up the sliders
interactive_plot = interactive(f, a=(-9, 9), zoom=(1,100))
interactive_plot
```
